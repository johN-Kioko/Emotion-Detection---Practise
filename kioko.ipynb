### Phase 1 - Loading The Dataset and Required Libraries

#### Importing Required Libraries

```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Loading The Dataset 

```python
# Loading the Dataset

data = pd.read_csv("Datasets/train.txt", sep=";", names=['Description', 'Emotion'])
data.head()
```

```python
data.info()
```

```python
data['Emotion'].value_counts()
```

### Phase 2 - EDA 

#### Mapping the emotions into numerical values

```python
# creating numerical values

data['Emotion_Numb'] = data['Emotion'].map({
    'joy' : 0,
    'sadness' : 1,
    'anger' : 2,
    'fear' : 3,
    'love' : 4,
    'surprise' : 5    
})

# confirming on the newly created numerical column
data.head()
```

#### Visualizations

```python
# Plotting a countplot of Emotions

plt.figure(figsize=(10,6))
sns.countplot(x='Emotion', data=data, hue='Emotion',palette='Set2', legend=False)
plt.title('Distribution of Categories')
plt.show()
```

```python
# Plotting a Histogram for Emotions

plt.figure(figsize=(10, 6))
sns.histplot(x='Emotion', data=data, kde=True, element='bars', stat='count', common_norm=False, color='skyblue')
plt.title('Distribution of Emotions')
plt.show()
```
